```python
# Data cleaning
```

```python
# Drop duplicates
```

```python
df = df.drop_duplicates
df
```

```python
# Drop columns
```

```python
df.drop(columns = "creativity")
```

```python
# Split name into first and second name 
```

```python
df[['fIrst_name' , 'second_name']] = df['Name'].str.split(' ',1, expand=True)
df
```

```python
# Convert data
```

```python
df['element_type'] = df['element_type'].str.replace('DEF', 'Defender')
df['element_type'] = df['element_type'].str.replace('FWD', 'Forward')
df['element_type'] = df['element_type'].str.replace('MID', 'Midfielder')
df['element_type'] = df['element_type'].str.replace('GK', 'Goalkeeper')
df
```

```python
# Rename column
```

```python
df = df.rename(columns = {'element_type': 'position'})
df
```

```python
# Strip spaces on either side of name
```

```python
df['Name'] = df['Name'].str.strip()
df
```

```python
# Exploratory data analysis
```

```python
df[df['total_points'] > 10].sort_values(by='total_points', ascending=False)
```

```python
df_sorted = df.sort_values(by='total_points', ascending=False)
df_sorted.head(10).plot(kind='bar', x='Name', y='total_points', legend=False, rot=45, figsize=(10, 6))
plt.xlabel('Player Names')
plt.ylabel('Total Points')
plt.title('Top 10 Players with Most Points')
plt.show()
```

```python
df.plot(kind='scatter', x='now_cost', y='total_points', color='orange', alpha=0.5, figsize=(10, 6))
plt.xlabel('Cost')
plt.ylabel('Total Points')
plt.title('Scatter Plot: Total Points vs. Cost')
plt.show()
```
